```python
#Question 3

def PrimeFinder(n):
    result = "Prime"
    if n <= 1:
        result = "Invalid Input"
    i = 2
    while i <= n/2:
        if n%i == 0:
            result = "Not Prime"
            break
        i = i + 1
    return result

print(PrimeFinder(31))
print(PrimeFinder(33))

x = 2
Primes = []
while x < 100:
    if PrimeFinder(x) == "Prime":
        Primes.append(x)
    x = x + 1
        
print(Primes)

def AllPrimesBelow(m):
    AllPrimes = []
    b = 2
    while b < m:
        if PrimeFinder(b) == "Prime":
            AllPrimes.append(b)
        b = b + 1
    results = AllPrimes
    return results

AllPrimesBelow(1000)
```

```python
#Question 4 part 1
import numpy as np
import random 

def NewArray(n, data):
    Array = np.empty((n, 3, 3))

    if data == "random":
        i = 0
        j = 0
        k = 0
        while i < n:
            while j < 3:
                while k < 3:
                    Array[i][j][k] = random.randint(1,100)
                    k = k + 1
                j = j + 1
                k = 0
            i = i + 1  
            j = 0
    else:
        i = 0
        j = 0
        k = 0
        while i < n:
            while j < 3:
                while k < 3:
                    Array[i][j][k] = data
                    k = k + 1
                j = j + 1
                k = 0
            i = i + 1  
            j = 0

    return Array

NewArray(10,0)
```

```python
#Question 6
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-5,5,.01)
y = np.ones(len(x))

a = 0
while a < len(x):
    y[a] = .1 * x[a] ** 3 - x[a] ** 2 + 5
    a = a + 1

a = 0
ymin = y[a]
ymax = y[a]
xmin = x[a]
xmax = x[a]
while a < len(x)-1:
    if y[a] < ymin:
        xmin = x[a]
        ymin = y[a]
    if y[a] > ymax:
        xmax = x[a]
        ymax = y[a]
    a = a + 1

plt.plot(x,y)
plt.plot(xmax, ymax, xmin, ymin, marker= 'x')
plt.legend(["Function","MAX","MIN"])
plt.show()
```
